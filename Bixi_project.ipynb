## Using a Neural Network to predict Bixi demand

```python
X_train=scaler.fit_transform(X_train)

X_train.shape
```

```python
X_test=scaler.transform(X_test)
```

```python
#BasicNeuralNetwork

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout,LSTM
from sklearn.metrics import r2_score, mean_squared_error
```

```python
#build model
from tensorflow.keras.callbacks import EarlyStopping
early_stop=EarlyStopping(monitor='val_loss',mode='min',verbose=1,patience=25)
model = Sequential()
model.add(Dense(23, activation='relu'))
# model.add(Dropout(0.2))
model.add(Dense(40, activation='relu'))
#model.add(Dropout(0.2))
model.add(Dense(15, activation='relu'))
#model.add(Dropout(0.2))
model.add(Dense(5, activation='relu'))
model.add(Dense(1,activation='linear'))
```

```python
model.compile(optimizer='adam',loss='mse')
```

```python
model.fit(X_train,y_train,epochs=200,validation_data=(X_test,y_test),batch_size=21,callbacks=[early_stop])
```

```python
loss_df=pd.DataFrame(model.history.history)
loss_df.plot()
```

```python
y_pred=model.predict(X_test)
```

```python
from sklearn.metrics import r2_score, accuracy_score, explained_variance_score, max_error, mean_squared_error
```

```python
np.sqrt(mean_squared_error(y_test,y_pred))
```

```python
explained_variance_score(y_test,y_pred)
```

```python
r2_score(y_test,y_pred)
```
